```python
def lowercase(data):
    data['word'] = data['word'].str.lower()
    return data
```

```python
lexi_liu = lowercase(lexi_liu)
```

```python
lexi_liu
```

```python
lexi_inset = lowercase(lexi_inset)
```

```python
lexi_liu.to_csv("lexiliu_final.csv" ,index = False,sep=";")
```
