## Klasifikasi

Klasifikasi yang dilakukan adalah dengan pendekatan Lexicon dengan library dari Bing Liu yang di traslate dalam bahasa Indonesia

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../data/dataset/datasetpraproses.csv", sep=';')
```

```python
dataset
```

```python
lexi = pd.read_csv("../data/lexicon/lexiconliu.csv", sep=";")
lexi
```

```python
def klasifikasi(text, lexicon):
    weight = 0
    for i in range(5):
        n = i+1
        n_grams = [str(text).split(' ')[i-(n-1):i+1] for i in range(n-1,len(str(text).split(' ')))]
        for j in n_grams:
            text_grams = ' '.join(j)
            for k in lexicon:
                if text_grams==k[0]:
                    weight = weight+k[1]
#     weight = ' '.join(map(str, text))
    return weight
```

```python
dataset['sentiment'] = dataset['Tweet_Parsed'].map(lambda com : klasifikasi(com,lexi.values))
```

```python
dataset
```

```python
dataset.to_csv("../data/klasifikasi/klasifikasi_liu.csv", sep=";", index=False)
```

# aspect extraction

```python
dataset = pd.read_csv("../data/klasifikasi/klasifikasi_liu.csv", sep=";")
```

```python
subaspect = pd.read_csv("../data/dataset/newsubaspect.csv", sep=";")
```

```python
subaspect
```

```python
lexi = pd.read_csv("../data/lexicon/lexiconliu.csv", sep=";")
```

```python
dataset
```

```python
label = []
for i in dataset['Tweet_Parsed']:
    split = i.split(" ")
    n = len(split)
    start = 0
    j = start
    sentiment = []
    while start<n and j<n:
        j = start
        score = 0
        if split[j] in subaspect['subaspect'].values:
            aspect = subaspect[subaspect['subaspect']==split[j]]['aspect'].values[0]
            if j==start:
                k = j+1
                while k<n:
                    if split[k] in subaspect['subaspect']:
                        break
                    k = k+1
                opinion = " ".join(split[j:k-1])
                score = klasifikasi(opinion,lexi.values)
                start = k
            elif j>start:
                k = j-1
                opinion = " ".join(split[start:k])
                score = klasifikasi(opinion,lexi.values)
                start = j
                if score==0 and j!=n-1:
                    k = j+1
                    while k<n:
                        if split[k] in subaspect['subaspect']:
                            break
                        k = k+1
                    opinion = " ".join(split[j:k-1])
                    score = klasifikasi(opinion,lexi.values)
                    start = k
#             score = "positif" if score>=0 else "negatif"
            if score>0:
                score = "positif"
            elif score==0:
                score = "netral"
            else:
                score = "negatif"
#             print(aspect)
            sentiment.append(" ".join([aspect,score]))
        else:
            start = start+1
#         print(subaspect['subaspect'])
#         print(sentiment)
    label.append(",".join(sentiment))
        
```
