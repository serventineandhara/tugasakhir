```python
from tesaurus_master.tesaurus import *
import pandas as pd
import numpy as np
```

```python
mydict = load("tesaurus_master/dict.json")
```

```python
subaspect = pd.read_csv("../data/dataset/aspect_subaspect.csv")
subaspect.head()
```

```python
new_subasp = []

for i in subaspect.values:
    asp = i[0]
    new_sub = getSinonim(i[1],mydict)
    if len(new_sub) > 0:
        for j in new_sub:
            new_subasp.append([asp,j])
```

```python
new_subasp
```

```python
new_subasp = pd.DataFrame(new_subasp,columns=['aspect','subaspect'])
```

```python
new_subasp.head()
```

```python
fix_subasp = subaspect.append(new_subasp)
fix_subasp.head()
```

```python
fix_subasp.to_csv("../data/dataset/newsubaspect_masihharusdisaring.csv",sep=';')
```
